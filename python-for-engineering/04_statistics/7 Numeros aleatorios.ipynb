# Numéros aleatorios

Van a haber ocasiones donde debamos simular valores que se extraen de forma aleatoria de alguna distribución. Esto es llamado *generación de números aleatorios*. Estos números aleatorios son llamados formalmente *pseudoaleatorios* ya que no están idealmente muestreados de una población pero están cerca de ello.

Una forma efectiva para testear la generación de números aleatorios es crear un histograma de los resultados y ver si coincide con la densidad de función.

Veamos un ejemplo para la distribución uniforme

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0, 1, 10000)
histvals, bin_edges = np.histogram(x)

bin_width = bin_edges[1]-bin_edges[0]
n = len(histvals)
bin_centers = np.zeros((n))
for i in range(n):
    bin_centers[i] = (bin_edges[i]+bin_edges[i+1])/2

plt.bar(bin_centers, histvals,width=bin_width,color='w', edgecolor='k')
plt.title('uniform random numbers')
```
