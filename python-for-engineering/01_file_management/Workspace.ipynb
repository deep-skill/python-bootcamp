```python
def calcular_promedio(lst):
    promedio = 0
    
    for nota in lst:
        promedio = promedio + nota
        
    promedio = promedio / len(lst)
    
    return promedio
```

```python
notas = []

with open('notas.txt', 'r') as fnotas:
    
    for j in range(1000):
    
        line = fnotas.readline()
        line = line[:-1]

        lst = line.split(' ')

        for i in range(len(lst)):
            lst[i] = int(lst[i])

        promedio = calcular_promedio(lst)
        
        lst.append(promedio)

        print(lst)
    
        notas.append(lst)
```

```python
print(notas)
```

```python
# Modo Escritura



with open('reporte.txt', 'w') as freporte:
    freporte.write('Nota1, Nota2, Nota3, Nota4, Promedio\n')
    
    for line in notas:
        for element in line:
            freporte.write(str(element))
            freporte.write(', ')
            
        freporte.write('\n')
```
